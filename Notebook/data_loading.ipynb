```python
df = pd.read_csv("Data/chatgpt.csv")
print(df.head())
```

```python
def overview_dataframe(df:pd.DataFrame):
  print("Dataframe Info")
  print(df.info())

  print("\n","*-*"*40,"\n")

  print("Dataframe Shape")
  print(f"Rows Shape: {df.shape[0]} and Column Shape: {df.shape[1]}")

  print("\n","*-*"*40,"\n")

  print("Data types")
  print(df.dtypes)

  print("\n","*-*"*40,"\n")

  print("Top 5 rows")
  print(df.head())

  print("\n","*-*"*40,"\n")

  print("Last 5 rows")
  print(df.tail())

  print("\n","*-*"*40,"\n")

  print("Summary Statistics")
  print(df.describe())

  print("\n","*-*"*40,"\n")

  print("Null Value Summary")
  print(df.isnull().sum())
```

```python
overview_dataframe(df)
```
